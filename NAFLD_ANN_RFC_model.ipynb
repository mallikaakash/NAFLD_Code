# **4) Artificial Neural Network Model**

```python
import tensorflow as tf

class ANNClassifier(tf.keras.Model):
    def __init__(self, input_size, output_size):
        super(ANNClassifier, self).__init__()
        
        self.model = tf.keras.Sequential()
        self.model.add(tf.keras.layers.Dense(input_size, activation='relu', input_shape=(input_size,)))
        self.model.add(tf.keras.layers.Dropout(0.5))
        self.model.add(tf.keras.layers.Dense(64, activation='tanh'))
        self.model.add(tf.keras.layers.Dropout(0.5))
        self.model.add(tf.keras.layers.Dense(32, activation='sigmoid'))
        self.model.add(tf.keras.layers.Dense(16))
        tf.keras.layers.Dense(output_size, activation='softmax') 

    def call(self, x, training=False):
        return self.model(x, training=training)
```

## **5) Autoencoder Code**

```python
# Encoder architecture
class Encoder(tf.keras.Model):
    def __init__(self, n_inputs, n_bottleneck):
        super(Encoder, self).__init__()

        # Encoder level 1
        self.layer1 = tf.keras.Sequential([
            layers.Dense(n_inputs * 2),
            layers.BatchNormalization(),
            layers.LeakyReLU()
        ])

        # Encoder level 2
        self.layer2 = tf.keras.Sequential([
            layers.Dense(n_inputs),
            layers.BatchNormalization(),
            layers.LeakyReLU()
        ])

        # Bottleneck
        self.bottleneck = layers.Dense(n_bottleneck)

    def call(self, x):
        x = self.layer1(x)
        x = self.layer2(x)
        x = self.bottleneck(x)
        return x


n_inputs = 29 
n_bottleneck = 7 

encoder = Encoder(n_inputs, n_bottleneck)

# Decoder architecture
class Decoder(tf.keras.Model):
    def __init__(self, n_bottleneck, n_inputs):
        super(Decoder, self).__init__()

        # Decoder level 1
        self.layer1 = tf.keras.Sequential([
            layers.Dense(n_inputs),
            layers.BatchNormalization(),
            layers.LeakyReLU()
        ])

        # Decoder level 2
        self.layer2 = tf.keras.Sequential([
            layers.Dense(n_inputs * 2),
            layers.BatchNormalization(),
            layers.LeakyReLU()
        ])

        # Output layer
        self.output_layer = layers.Dense(n_inputs)

    def call(self, x):
        x = self.layer1(x)
        x = self.layer2(x)
        x = self.output_layer(x)
        return x


decoder = Decoder(n_bottleneck, n_inputs)

# Combine Encoder and Decoder to create the Autoencoder
class Autoencoder(tf.keras.Model):
    def __init__(self, encoder, decoder):
        super(Autoencoder, self).__init__()
        self.encoder = encoder
        self.decoder = decoder

    def call(self, x):
        x = self.encoder(x)
        x = self.decoder(x)
        return x




```

```python
def reduce_dimensionality(data, num_output_features, autoencoder):
    encoder = autoencoder.encoder
    
    reduced_features = encoder(data)
    
    reduced_features = layers.Dense(num_output_features, activation='relu')(reduced_features)
    
    return reduced_features
```
